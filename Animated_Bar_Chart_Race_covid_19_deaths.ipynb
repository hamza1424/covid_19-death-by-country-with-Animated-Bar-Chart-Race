```python
# Selecting Countries with High Cases

cols = ['date', 'Italy', 'Australia', 'Brazil', 'India', 'Brazil', 'India', 'United Kingdom', 'Pakistan',
       'Spain', 'Canada', 'New Zealand', 'Russia', 'Thailand']
```

```python
# Data Processing
data_subset = data[cols]

data_subset.set_index('date', inplace = True)

cum_sum_data = data_subset.cumsum(axis = 0)

cum_sum_data.tail(10)
```

```python
# Using Bar Chart Race
import bar_chart_race as bcr

bcr.bar_chart_race(df = cum_sum_data, filename=None, figsize=(5,3), title = 'COVID-19 Cases By Country')
```
